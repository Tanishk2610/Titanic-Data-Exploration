# Assignment 1: Write a program to scheme a few activation functions that are used in neural networks

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
    Sigmoid activation function.
    """
    return 1 / (1 + np.exp(-x))

def relu(x):
    """
    Rectified Linear Unit (ReLU) activation function.
    """
    return np.maximum(0, x)

def tanh(x):
    """
    Hyperbolic Tangent (tanh) activation function.
    """
    return np.tanh(x)

def softmax(x):
    """
    Softmax activation function for output layer in multiclass classification.
    """
    exp_vals = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_vals / np.sum(exp_vals, axis=-1, keepdims=True)

# Generate x values
x = np.linspace(-5, 5, 100)

# Compute activation function values
sigmoid_y = sigmoid(x)
relu_y = relu(x)
tanh_y = tanh(x)
softmax_y = softmax(x)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, sigmoid_y, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, relu_y, label='ReLU', color='orange')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, tanh_y, label='Tanh', color='green')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, softmax_y, label='Softmax', color='red')
plt.title('Softmax Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    """
    Sigmoid activation function.
    """
    return 1 / (1 + np.exp(-x))

def relu(x):
    """
    Rectified Linear Unit (ReLU) activation function.
    """
    return np.maximum(0, x)

def tanh(x):
    """
    Hyperbolic Tangent (tanh) activation function.
    """
    return np.tanh(x)

def softmax(x):
    """
    Softmax activation function for output layer in multiclass classification.
    """
    exp_vals = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_vals / np.sum(exp_vals, axis=-1, keepdims=True)

def linear(x):
    """
    Linear activation function.
    """
    return x

# Generate x values
x = np.linspace(-5, 5, 100)

# Compute activation function values
sigmoid_y = sigmoid(x)
relu_y = relu(x)
tanh_y = tanh(x)
softmax_y = softmax(x)
linear_y = linear(x)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, sigmoid_y, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 3, 2)
plt.plot(x, relu_y, label='ReLU', color='orange')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 3, 3)
plt.plot(x, tanh_y, label='Tanh', color='green')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 3, 4)
plt.plot(x, softmax_y, label='Softmax', color='red')
plt.title('Softmax Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(x, linear_y, label='Linear', color='purple')
plt.title('Linear Activation Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```
